```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import pandas as pd
import numpy as n
df = pd.read_excel('Clean_DF.xlsx')
#df_test = pd.DataFrame([{'count':df['bhk'].value_counts(),'type':df['bhk'].unique()}]columns=['type','count'])

#df_test = pd.DataFrame(type_flat,count)

print(dcc.__version__) # 0.6.0 or above is required

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

# Since we're adding callbacks to elements that don't exist in the app.layout,
# Dash will raise an exception to warn us that we might be
# doing something wrong.
# In this case, we're adding the elements through a callback, so we can ignore
# the exception.
app.config.suppress_callback_exceptions = True

app.layout = html.Div([
    dcc.Location(id='url', refresh=False),
    html.Div(id='page-content')
])


index_page = html.Div([
    html.Center(html.H1('EDA')),
    html.Center(html.P('Exploratory Data Analysis of the data scraped from Magicbricks.com which will help in understanding which features help in the growth of Retail Property Price')),
    html.Center(dcc.Link('Go to Page 1 of EDA', href='/page-1')),
    html.Br(),
    html.Center(dcc.Link('Go to Page 2 of EDA', href='/page-2')),
    html.Br(),
    html.Center(dcc.Link('Go to Page 3 of EDA', href='/page-3')),
])

page_1_layout = html.Div([
    html.Div([
        html.Div([
            html.H3('Floor V/s Price'),
            dcc.Graph(id='g1', figure={'data': [go.Scatter(
                x = df['floor'],
                y =df['price in lakhs'],
                mode='markers',
                opacity=0.7,
                marker={
                        'size': 15,
                        'line': {'width': 0.5, 'color': 'white'}
                    },
 
            )],
                                            
                                           
           'layout': go.Layout(
                xaxis={'title': 'Floor'},
                yaxis={'title': 'Price in Lakhs'},
                #margin={'l': 40, 'b': 40, 't': 10, 'r': 10},
                #legend={'x': 0, 'y': 1},
                hovermode='closest'
            )
                                      
                                      
                                      
                                      })
        ], className="six columns"),

        html.Div([
            html.H3('BHK V/s Price'),
            dcc.Graph(id='g2', figure={'data': [{'x': df['bhk'],
            'y': df['price in lakhs'],
            'type': 'bar'}],
               
             'layout': {
                
                'xaxis':{'title': 'BHK'},
                'yaxis':{'title': 'Price in Lakhs'},
            }    
                                      
                                      
                                      
                                      
                                      
                                      
                                      
                                      })
        ], className="six columns"),
    ], className="row"),

    html.Div(id='page-1-content'),
    html.Br(),
    dcc.Link('Go to Page 2', href='/page-2'),
    html.Br(),
    dcc.Link('Go to Page 3', href='/page-3'),
    html.Br(),
    dcc.Link('Go back to home', href='/'),
])



page_2_layout = html.Div([
    html.Div([
        html.Div([
            html.H3('Facing V/s Price'),
            dcc.Graph(id='g1', figure={'data': [{'x': df['facing'],
            'y': df['price in lakhs'],
            'type': 'bar'}],
                                       
               'layout': {
                
                'xaxis':{'title': 'House Facing'},
                'yaxis':{'title': 'Price in Lakhs'},
            }                          
                                      
                                      
                                      
                                      
                                      })
        ], className="six columns"),

        html.Div([
            html.H3('Transaction V/s Price'),
            dcc.Graph(id='g2', figure={'data': [{'x': df['transaction'],
            'y': df['price in lakhs'],
            'type': 'bar'}],
                                      
              'layout': {
                
                'xaxis':{'title': 'Transaction Type'},
                'yaxis':{'title': 'Price in Lakhs'},
            }                                   
                                      
                                      })
        ], className="six columns"),
        
       
        
        
    ], className="row"),
    
    html.Div(id='page-2-content'),
    html.Br(),
    dcc.Link('Go to Page 1', href='/page-1'),
    html.Br(),
    dcc.Link('Go to Page 3', href='/page-3'),
    html.Br(),
    dcc.Link('Go back to home', href='/'),
])


page_3_layout = html.Div([
    html.Div([
        html.Div([
            html.H3('Total Area v/s Price'),
            dcc.Graph(id='scatter', figure={'data': [go.Scatter(
                x = df['total area'],
                y =df['price in lakhs'],
                mode='markers',
                opacity=0.7,
                marker={
                        'size': 15,
                        'line': {'width': 0.5, 'color': 'white'}
                    },
 
            )],
                                            
                                           
           'layout': go.Layout(
                xaxis={'title': 'Total Area'},
                yaxis={'title': 'Price in Lakhs'},
                #margin={'l': 40, 'b': 40, 't': 10, 'r': 10},
                #legend={'x': 0, 'y': 1},
                hovermode='closest'
            )
                                           
                                           })
        ], className="six columns"),

        html.Div([
            html.H3('Furnishing V/s Price'),
            dcc.Graph(id='g2', figure={'data': [{'x': df['furnishing'],
            'y': df['price in lakhs'],
            'type': 'bar'}],
            'layout': {
                
                'xaxis':{'title': 'Furnishing'},
                'yaxis':{'title': 'Price in Lakhs'},
            }    

                                      
                                      
                                      })
        ], className="six columns"),
        
       
        
        
    ], className="row"),
    
    html.Div(id='page-3-content'),
    html.Br(),
    dcc.Link('Go to Page 1', href='/page-1'),
    html.Br(),
    dcc.Link('Go to Page 2', href='/page-2'),
    html.Br(),
    dcc.Link('Go back to home', href='/'),
])




# Update the index
@app.callback(dash.dependencies.Output('page-content', 'children'),
              [dash.dependencies.Input('url', 'pathname')])
def display_page(pathname):
    if pathname == '/page-1':
        return page_1_layout
    elif pathname == '/page-2':
        return page_2_layout
    elif pathname == '/page-3':
        return page_3_layout
    else:
        return index_page
    # You could also return a 404 "URL not found" page here


if __name__ == '__main__':
    app.run_server()
```
